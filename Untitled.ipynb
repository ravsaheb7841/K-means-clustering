```python
# Import Required Libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
# Load the Dataset

data = pd.read_csv("Mall_Customers.csv")
```

```python
data.head()
```

```python
# Information of the data

data.info()
```

```python
# Check the null value in the data

data.isnull().sum()
```

```python
data.describe()
```

```python
#  Select Features for Clustering

x = data[["CustomerID", "Age", "Annual Income (k$)", "Spending Score (1-100)"]]
```

```python
# Apply StandardScaler

sc = StandardScaler()
```

```python
x_scaler = sc.fit_transform(x)
```

```python
# Finding the Optimal Number of Clusters Using the Elbow Method

inertia = []
k_range = range(1, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(x_scaler)
    inertia.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(k_range, inertia, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
# Apply K-Means Clustering with the Optimal Number of Clusters

k_optimal = 5
kmeans = KMeans(n_clusters=k_optimal, random_state=0)
data['Cluster'] = kmeans.fit_predict(x_scaler)
```

```python
# Visualize the Clusters

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cluster', palette='viridis', data=data, style='Cluster', markers=['o', 's', 'D', '^', 'v'])
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label='Centroids', marker='X')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Customer Segments Based on Annual Income and Spending Score')
plt.legend(title='Cluster')
plt.show()
```
